### import dataset and extract features and targets
https://archive.ics.uci.edu/ml/machine-learning-databases/iris/


```python
import pandas as pd

url = "https://archive.ics.uci.edu/ml/machine-learning-databases/iris/iris.data"
#the imported dataset does not have the required column names so lets add it
colums = ['sepal length (cm)', 'sepal width (cm)', 'petal length (cm)', 'petal width (cm)', 'Class']

data = pd.read_csv(url, names=colums)

features = data.iloc[:,0:4].values
target = data.iloc[:,-1].values
data
```

### construct a decision tree


```python
from sklearn import tree

clf = tree.DecisionTreeClassifier()
clf = clf.fit(features, target)
```

### Lets export the tree in Graphviz format using the export_graphviz exporter.

```python
import graphviz

feature_names=['sepal-length', 'sepal-width', 'petal-length', 'petal-width']
target_names = ['setosa', 'versicolor' ,'virginica']

dot_data = tree.export_graphviz(clf, 
                                out_file=None,
                                feature_names=feature_names,
                                class_names=target_names, 
                                filled=True, 
                                rounded=True,  
                                special_characters=True)

graph = graphviz.Source(dot_data) 
graph
```

After being fitted, the model can then be used to predict the class of samples:

```python
from sklearn.metrics import classification_report
print(classification_report(clf.predict(features), target))
```
