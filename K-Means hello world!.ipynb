### Generate a random dataset with 4 clusters
we will be applying k-means on this data and see how accurate out prediction is.

```python
from sklearn.datasets.samples_generator import make_blobs
import matplotlib.pyplot as plt

random_state = 0
clusters = 4
features, target = make_blobs(n_samples=300, centers=clusters, cluster_std=0.60, random_state=random_state)

#plot the data to see how it looks 
plt.scatter(features[:,0], features[:,1])
```

### Apply K-Means

```python
from sklearn.cluster import KMeans

kmeans = KMeans(n_clusters=clusters,  random_state=random_state)#which is 4
kmeans = kmeans.fit(features) #modal building
y_predicted = kmeans.predict(features) #prediction
```

### Plot the analysis

```python
print kmeans.cluster_centers_
colors = ['r','g','b','y']

predicted_colors = []
for t in y_predicted:
    predicted_colors.append(colors[t])
    
plt.scatter(features[:,0], features[:,1],  c=predicted_colors)
plt.scatter(kmeans.cluster_centers_[:,0], kmeans.cluster_centers_[:,1], s=5000, c=colors,  alpha=0.2)
```
